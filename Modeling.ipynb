```python
from sklearn.preprocessing import StandardScaler
```

```python
stdsc = StandardScaler()
X_std = stdsc.fit_transform(df[cols].iloc[:, range(0,20)].values)

cov_mat = np.cov(X_std.T)
plt.figure(figsize=(40,40))
sns.set(font_scale=2.0)
hm = sns.heatmap(cov_mat,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size':14},
                 cmap='coolwarm',
                 yticklabels=cols,
                 xticklabels=cols)
plt.title('Covariance Matrix showing Correlation Coefficients', size=50)
plt.tight_layout()
plt.savefig('coefficients.jpeg')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()

feature_cols = ['mrp', 'retailer_amz', 'retailer_aeus', 'review_count']

X = df[feature_cols]
y = df.price
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
round(mean_squared_error(y_test, y_pred),2)
```

```python
round(mean_absolute_error(y_test, y_pred))
```

Mean absolute error seems more fair, given that being off by 10 is not more than twice as bad as being off by 5.
